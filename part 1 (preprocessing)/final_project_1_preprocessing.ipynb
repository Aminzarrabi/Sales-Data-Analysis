```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('sales.xlsx')
df
```

```python
df_no_missing_CustomerId = df.copy()
# df_no_missing_CustomerId.dropna('CustomerId')


df_no_missing_CustomerId = df_no_missing_CustomerId.drop(df_no_missing_CustomerId.loc[df_no_missing_CustomerId.CustomerId.isna()].index)
df_no_missing_CustomerId
```

```python
df_no_duplicate = df_no_missing_CustomerId.copy()

df_no_duplicate = df_no_duplicate.drop_duplicates()

df_no_duplicate
```

```python
df_no_missing_price = df_no_duplicate.copy()
# df_no_missing_CustomerId = df_no_missing_CustomerId.drop(loc[df_no_missing_CustomerId.CustomerId.isna()].index)
df_no_missing_price = df_no_missing_price.drop(df_no_missing_price.loc[df_no_missing_price['UnitPrice'] <= 0].index)
df_no_missing_price
```

```python
dff = df_no_missing_price.copy()
# dff['InvoiceNumber'] = dff['InvoiceNumber'].astype('string')
# x = dff[dff['InvoiceNumber'].str.contains("C")]
# y = dff['InvoiceNumber'].value_counts().sum()
# percentage_cancelled_orders = x
# percentage_cancelled_orders

canceled_orders = dff[dff['InvoiceNumber'].str.startswith('C', na=False)]

total_orders = dff['InvoiceNumber'].nunique()

num_canceled_orders = canceled_orders['InvoiceNumber'].nunique()


percentage_cancelled_orders =round((num_canceled_orders / total_orders) * 100)
 
percentage_cancelled_orders
```

```python

# top_cancelling_customers = None
# d = df_no_missing_price.groupby('CustomerId')['InvoiceNumber'].apply(lambda x : x.str.startswith('C', na=False))
d = canceled_orders.groupby('CustomerId')['InvoiceNumber'].nunique().sort_values(ascending = False).index[0:5]
d = d.astype('int32')
d = d.astype('string').to_list()
top_cancelling_customers = d
top_cancelling_customers 
```

```python
df_no_canceled_invoice = df_no_missing_price.copy()
canceled_orders = df[df['InvoiceNumber'].str.startswith('C', na=False)]
# df_no_canceled_invoice = df_no_canceled_invoice.drop(df_no_canceled_invoice.index[canceled_orders])
df_no_canceled_invoice = pd.concat([df_no_canceled_invoice, canceled_orders, canceled_orders]).drop_duplicates(keep=False)
df_no_canceled_invoice
# df1[df1.apply(lambda x: x.values.tolist() not in df2.values.tolist(), axis=1)]
df_no_canceled_invoice
```
